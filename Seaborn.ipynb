# Seaborn
> - for visual data
> - sample dataset in seaborn
> - better than matplot less lines of code

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
irs = sns.load_dataset('iris')
# print(irs.values)
print(irs.species.value_counts())
print(irs.shape)
irs
```

# Lineplot
> see the trend in the data

```python
sns.lineplot(x='petal_length',y='petal_width',data=irs) 
plt.show()
```

```python
sns.lineplot(x='petal_length',y='petal_width',data=irs,errorbar=None)  # can use errorbar or ci
plt.show()
```

# Scatter Plots
> relationship between 2 variables

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=irs,hue='species') # hue will give color based on species of the data(in here only)
plt.title("Scatter Plot")
plt.show()
```

# Bar Chart
> compare

```python
sns.barplot(x='species',y='petal_length',data=irs,hue="species")
plt.title("Bar Chart")
plt.show()
```

# Count Plot
A count plot in Seaborn is used to show the counts of observations in each categorical bin using bars. It's especially useful for visualizing the distribution of a categorical variable.

```python
sns.countplot(x="species",data=irs,hue="species")
plt.show()
```

# Boxplot
A boxplot in Seaborn (and matplotlib) is used to visualize the distribution, central value (median), and variability (IQR) of a dataset. It also shows potential outliers.
> A boxplot summarizes the distribution of a dataset using five key values:
> - Minimum (excluding outliers)
> - 1st Quartile (Q1) â€“ 25th percentile
> - Median (Q2) â€“ 50th percentile
> - 3rd Quartile (Q3) â€“ 75th percentile
> - Maximum (excluding outliers)

```python
sns.boxplot(x="sepal_width",data=irs)
plt.show()
```

```python
sns.boxplot(x="species",y="sepal_width",data=irs,hue="species")
plt.show()
```

# Histogram
> frequency
<br> like bar chart but bars are connected

| Parameter          | Meaning                                                                       |
| ------------------ | ----------------------------------------------------------------------------- |
| `x="sepal_width"`  | Tells Seaborn which column to plot on the X-axis (histogram of `sepal_width`) |
| `data=irs`         | The DataFrame (likely `iris` dataset, shortened as `irs`)                     |
| `bins=15`          | Divides the range of data into 15 intervals (bars)                            |
| `stat="frequency"` | The height of each bar shows how many data points fall in each bin            |
| `color="red"`      | Sets the color of the histogram bars                                          |
| `kde=True`         | Adds a **KDE** (Kernel Density Estimate) line to smooth the distribution      |


```python
sns.histplot(x="sepal_width",data=irs,bins=15,stat="frequency",color="yellow")
```

```python
sns.histplot(x="sepal_width",data=irs,bins=15,stat="frequency",color="red",kde=True)
```

# KDE plot (Kernel Density Estimate)
> smooth curve that estimates the probability distribution of a continuous variable. Itâ€™s an alternative to histograms but gives a more elegant and continuous view of the data distribution.



```python
sns.kdeplot(x="sepal_length",data=irs,color="red",fill=True)
```
